```python
import pandas as pd
import re
```

```python
#download from http://public.synergylab.cn/dcdb/download.jsf
```

```python
dc = pd.read_csv("drugCombinations.txt",sep='\t')
identifier = pd.read_csv("components_identifier.txt",sep='\t')
dc2dcu = pd.read_csv("DC_TO_DCU.txt",sep='\t')
dc_usage = pd.read_csv("DC_USAGE.txt",sep='\t')
# identifier
```

```python
#dc to dcu
_dc2dcu = pd.merge(dc,dc2dcu,left_on='DrugCombination_ID',right_on='DC_ID')
# dcu to usage
_dc2dcu2usage = pd.merge(_dc2dcu,dc_usage,on='DCU_ID')
_dc2dcu2usage
```

```python
set(_dc2dcu2usage['DISEASE'])
```

```python
_dc2dcu2usage[_dc2dcu2usage['DISEASE'].isin(['Lymphoma','Lymphoma and leuckemia'\
                                            'Lymphoma, Low-Grade','Lymphoma, Mucosa-Associated Lymphoid Tissue',\
                                             'Lymphoma, Non-Hodgkin', "Non-Hodgkin's Lymphoma",\
                                            'Leukaemia, Lymphocytic, Chronic'])]
```

```python
_dc = dc.Componets_ID.str.split('/',2, expand=True)
_dc2 = _dc[_dc.iloc[:,2].isna()]
_dc2.columns = ['drug1','drug2','3']
_dc2 = _dc2.iloc[:,:-1]
```

```python
# define conversion dictionary to drugbank ID
_identifier = identifier[identifier.DrugBank.notna()]
conversion = _identifier[['DCC_ID','DrugBank']]
```

```python
_dc2_m1 = pd.merge(_dc2,conversion,left_on='drug1',right_on='DCC_ID')
_dc2_m2 = pd.merge(_dc2_m1,conversion,left_on='drug2',right_on='DCC_ID')
_dc2_m = _dc2_m2[['DrugBank_x','DrugBank_y']]
```

```python
_dc2_m
```

```python
def split_it(compound):
    return int(re.split('\d*\D+',compound)[1])
```

```python
_dc2_m['DrugBank_x'] = _dc2_m['DrugBank_x'].apply(lambda x: split_it(x))
_dc2_m['DrugBank_y'] = _dc2_m['DrugBank_y'].apply(lambda x: split_it(x))
```

```python
_dc2_m['cell'] = "nan"
_dc2_m['score'] = 1
```

```python
_dc2_m.to_csv("clinicaltrials.csv")
```
